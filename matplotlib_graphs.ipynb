```python
from scipy import signal
from matplotlib.pyplot import figure

a = signal.unit_impulse(11, 5)
n = np.linspace(-5, 5, 11)


figure(num=None, figsize=(4, 3), dpi=150, facecolor='w', edgecolor='k')

plt.stem(n, a)
```
